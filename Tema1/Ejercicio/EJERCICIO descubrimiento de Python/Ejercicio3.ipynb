#Ejercicio 3: Cálculo de Áreas

1.  Crea una nueva hoja de Colab.
2.  Escribe un programa en Python que calcule el área de diversas formas geométricas, como cuadrados, rectángulos y círculos.
3.  Solicita al usuario que elija la forma,ponga los datos necesarios (longitud de lados, radio, etc.) y muestra el área correspondiente.
4.  Ejecuta el programa y verifica que los cálculos de áreas sean correctos.


```python
import math

# Función para calcular el área de un cuadrado
def calcular_area_cuadrado(lado):
    return lado ** 2

# Función para calcular el área de un rectángulo
def calcular_area_rectangulo(ancho, largo):
    return ancho * largo

# Función para calcular el área de un círculo
def calcular_area_circulo(radio):
    return math.pi * (radio ** 2)

# Opciones de formas geométricas
opciones = {1: "cuadrado", 2: "rectangulo", 3: "circulo"}

# Solicitar al usuario que elija la forma
print("Elige una forma geométrica:")
for key, value in opciones.items():
    print(f"{key}: {value}")

opcion = int(input("Ingresa el número correspondiente a la forma: "))

# Calcular el área basado en la forma seleccionada
if opcion == 1:
    lado = float(input("Ingresa la longitud de un lado del cuadrado: "))
    area = calcular_area_cuadrado(lado)
elif opcion == 2:
    ancho = float(input("Ingresa el ancho del rectángulo: "))
    largo = float(input("Ingresa el largo del rectángulo: "))
    area = calcular_area_rectangulo(ancho, largo)
elif opcion == 3:
    radio = float(input("Ingresa el radio del círculo: "))
    area = calcular_area_circulo(radio)
else:
    print("Opción no válida. Por favor, elige una opción válida.")

# Mostrar el resultado
if opcion in [1, 2, 3]:
    print(f"El área del {opciones[opcion]} es {area:.2f}")
```
