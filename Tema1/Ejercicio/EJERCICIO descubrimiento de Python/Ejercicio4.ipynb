#Ejercicio 4: Listas y Bucles

1.  Crea una nueva hoja de Colab.
2.  Escribe un programa en Python que solicite al usuario una lista de números separados por comas.
3.  Convierte la entrada del usuario en una lista y realiza operaciones como sumar los números, encontrar el número más grande o más pequeño, y calcular el promedio.
4.  Ejecuta el programa y verifica que los cálculos sean precisos.


```python
# Solicitar al usuario una lista de números separados por comas
entrada = input("Ingresa una lista de números separados por comas: ")

# Dividir la entrada en una lista de números
numeros = [float(x.strip()) for x in entrada.split(',')]

# Verificar si se ingresaron números
if not numeros:
    print("No se ingresaron números válidos.")
else:
    # Calcular la suma de los números
    suma = sum(numeros)

    # Encontrar el número más grande y el más pequeño
    maximo = max(numeros)
    minimo = min(numeros)

    # Calcular el promedio
    promedio = suma / len(numeros)

    # Mostrar los resultados
    print(f"Suma de los números: {suma}")
    print(f"Número más grande: {maximo}")
    print(f"Número más pequeño: {minimo}")
    print(f"Promedio de los números: {promedio:.2f}")
```
