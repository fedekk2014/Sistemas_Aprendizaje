#Ejercicio 2: Conversión de Unidades

1.  Crea una nueva hoja de Colab.
2.  Escribe un programa en Python que permita al usuario convertir entre unidades de medida, como de grados Celsius a Fahrenheit o de kilómetros a millas.
3.  Solicita al usuario un valor y la unidad de medida original, y luego realiza la conversión y muestra el resultado.
4.  Ejecuta el programa y verifica que las conversiones sean precisas.


```python
def celsius_a_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

def kilometros_a_millas(kilometros):
    millas = kilometros * 0.621371
    return millas

while True:
    print("Elija una opción de conversión:")
    print("1. Convertir de grados Celsius a grados Fahrenheit")
    print("2. Convertir de kilómetros a millas")
    print("3. Salir")

    opcion = input("Ingrese el número de la opción deseada: ")

    if opcion == "1":
        celsius = float(input("Ingrese la temperatura en grados Celsius: "))
        fahrenheit = celsius_a_fahrenheit(celsius)
        print(f"{celsius} grados Celsius son equivalentes a {fahrenheit} grados Fahrenheit.")
    elif opcion == "2":
        kilometros = float(input("Ingrese la distancia en kilómetros: "))
        millas = kilometros_a_millas(kilometros)
        print(f"{kilometros} kilómetros son equivalentes a {millas} millas.")
    elif opcion == "3":
        print("¡Hasta luego!")
        break
    else:
        print("Opción no válida. Por favor, seleccione una opción válida.")
```
