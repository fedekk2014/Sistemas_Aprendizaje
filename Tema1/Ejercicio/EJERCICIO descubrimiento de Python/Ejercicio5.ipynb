#Ejercicio 5: Análisis de Texto

1.  Crea una nueva hoja de Colab.
2.  Escribe un programa en Python que solicite al usuario un párrafo de texto.
3.  Realiza análisis del texto, como contar las palabras, contar las letras y encontrar la palabra más larga.
4.  Ejecuta el programa y verifica que los análisis del texto sean correctos.


```python
# Solicitar al usuario un párrafo de texto
texto = input("Ingresa un párrafo de texto: ")

# Contar las palabras en el texto
palabras = texto.split()
numero_palabras = len(palabras)

# Contar las letras en el texto
letras = ''.join(palabras)
numero_letras = len(letras)

# Encontrar la palabra más larga
palabra_mas_larga = max(palabras, key=len)

# Mostrar los resultados
print(f"Número de palabras: {numero_palabras}")
print(f"Número de letras: {numero_letras}")
print(f"Palabra más larga: '{palabra_mas_larga}' con {len(palabra_mas_larga)} letras.")
```
